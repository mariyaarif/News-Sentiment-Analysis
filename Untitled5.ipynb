```python
import requests
import pandas as pd
from textblob import TextBlob 
```

```python
api_key = "YOUR_API_KEY_HERE"
url = f'https://newsapi.org/v2/everything?q=Artificial%20Intelligence&apiKey={api_key}'
reponse = requests.get(url)
data = reponse.json()
print(data['articles'][0]['title'])
```

```python
my_list = [] 

for article in data['articles']:
    title = article['title']
    analysis = TextBlob(title)
    score = analysis.sentiment.polarity
    my_list.append({"Headline" : title , "Score" : score})


df = pd.DataFrame(my_list)
print(df.head())
```

```python
def semtiment_analysis(score):
    if score < 0 : return " Negative"
    elif score > 0 : return "Positive"
    else : return  "Neutral"

df['Sentiment'] = df['Score'].apply(semtiment_analysis)
print(df)
```

```python
import matplotlib.pyplot as plt

df['Sentiment'].value_counts().plot(kind="bar" , color = ['green' , 'blue' , 'orange'])
plt.title("News Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Number of Headline")
plt.show()
```

```python
df.to_csv("mysentimrnt analysis.csv" , index=False)
print("Successfully saved")
```
